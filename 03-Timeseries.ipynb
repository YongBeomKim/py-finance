## **1 시계열 데이터 분해**
### **01 Loading the DataSet**
- 'GC=F' : Gold Price Code

````python
# Loading the DataSet
import os
import pandas as pd
file_name = './data/stock-amazon.csv'

if os.path.isfile(file_name): # Checking the File exist
    data = pd.read_csv(file_name)
    data['Date'] = pd.to_datetime(data['Date']) # datetime 포맷 변경
    data = data.set_index('Date')               # 변경된 컬럼을 Index 로 설정
else:  # if not, download the data
    import yfinance as yf
    data = yf.download('AMZN', start='2010-01-01', end='2021-03-01',
        progress=False, auto_adjust=True).reset_index()
    data.to_csv(file_name, index=None)
    data = data.set_index('Date')   
data.head(3)
````

### **02 Data Pre Processing**

````python
# Data Pre Processing
data.rename(columns={'Close':'price'}, inplace=True)
df = data[['price']]
df = df.resample('M').last()  # 월간 데이터로 변환하기

# 이동평균과 표준편차를 추가 한다
WINDOW_SIZE = 12
df['rolling_mean'] = df.price.rolling(window=WINDOW_SIZE).mean()
df['rolling_std'] = df.price.rolling(window=WINDOW_SIZE).std()
````

### **04 승산적 모델을 활용하여 계절성 특징을 찾는다**

````python
from statsmodels.tsa.seasonal import seasonal_decompose

plt.rcParams['figure.figsize'] = [12, 6]
decomposition_results = seasonal_decompose(df.price, model='multiplicative')
decomposition_results.plot().suptitle('Multiplicative Decomposition', fontsize=18)
plt.show()
````
